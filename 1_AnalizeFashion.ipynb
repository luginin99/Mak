```python
import numpy as np
from matplotlib import pyplot as plt
import random
%matplotlib inline
```

```python
def read_images(file_name):
    number_of_images, size_x, size_y = np.fromfile(file = file_name, dtype = np.uint32, count = 4).byteswap(inplace = True)[1:4]
    data = np.fromfile(file = file_name, dtype = np.uint8).byteswap(inplace = True)[16:].reshape((number_of_images, size_x, size_y)).astype(float)
    return data, number_of_images, size_x, size_y

def read_labels(file_name, number_of_labels):
    return np.fromfile(file = file_name, dtype = np.uint8).byteswap(inplace = True)[8:].reshape((number_of_labels))
```

```python
file_train_images = 'train-images-idx3-ubyte'
file_train_labels = 'train-labels-idx1-ubyte'
file_test_images  = 't10k-images-idx3-ubyte'
file_test_labels  = 't10k-labels-idx1-ubyte'

train_images, number_of_images, size_x, size_y = read_images(file_train_images)

train_labels = read_labels (file_train_labels, number_of_images)

test_images, number_of_test_images, size_x, size_y = read_images(file_test_images)

test_labels = read_labels (file_test_labels, number_of_test_images)
```

```python
fig=plt.figure(figsize = (size_x, size_y))
columns = 3
rows = 1
for i in range(1, 4):
    a = random.randint(0,number_of_images)
    fig.add_subplot(rows, columns, i)
    plt.imshow(train_images[a], cmap=plt.get_cmap('gray'))
plt.show()
```
